```python
df['date_added']=pd.to_datetime(df['date_added'],errors='coerce')
df['year_added']=df['date_added'].dt.year
df

```

```python
df['year_added'].value_counts().sort_index().plot(kind='bar', color='red')
plt.title('Titles added to netflix over a year')
plt.xlabel('year_added')
plt.ylabel('title')
plt.show()
```

#genres analysis

```python
from collections import Counter
```

```python
genre_series=df['listed_in'].dropna().str.split(', ')
all_genre=[genre for sublist in genre_series for genre in sublist]
genre_count=pd.Series(Counter(all_genre)).sort_values(ascending=False)
genre_count.head(10).plot(kind='bar',color='red')
plt.title('top 10 genre in netflix')
plt.show()
```
